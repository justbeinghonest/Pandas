```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r'NDAP_REPORT_7150.csv').dropna()
df.replace(np.nan,0)
plt.figure(figsize=[15,15])
plt.xlabel('State')
plt.ylabel('Pending Case (Count)')
Pending_cases=pd.pivot_table(df,index=['State']
                             ,values=['Pending cases','Original pending cases']
                             ,aggfunc=np.sum)
sort=Pending_cases.sort_values(by='Pending cases',ascending=False).plot(kind='barh')
plt.show()
print(Pending_cases.head(10))
```
