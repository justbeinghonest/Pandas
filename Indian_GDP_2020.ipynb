```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv(r'India_GDP_1960-2022.csv')

fig,ax = plt.subplots(2,2)

ax[0,0].bar(df['Year'],df['Growth %'],color='red')
ax[0,1].bar(df['Year'],df['GDP in (Billion) $'],color='blue')
ax[1,0].plot(df['Year'],df['Per Capita in rupees'],color='green')
plt.show()

print(df.head())
```
