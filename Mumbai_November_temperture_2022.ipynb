```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r'Temp_Mum November.csv')

df[['Celsius','Ferhanite']] = df['Temperature (¡C) | (¡F)'].str.split("|", expand = True)
df['New_Date'] = df['Date'].replace(regex=['. November'],value='').astype(np.int32)
col_cel = df['Celsius'].replace(regex=[' ¡C '],value='').astype({'Celsius':'float'}).dropna()
col_feh = df['Ferhanite'].replace(regex=[' ¡F'],value='').astype({'Ferhanite':'float'}).dropna()

df[['Milimeter','Inches']] = df['Precipitation / Rainfall (mm) | (inch.)'].str.split("|", expand = True)
col_mil = df['Milimeter'].replace(regex=[' mm '],value='').dropna().astype({'Milimeter':'float'})
col_inch = df['Inches'].replace(regex=[' inch.'],value='').dropna().astype({'Inches':'float'})


fig,ax = plt.subplots(2,1,figsize=(20,10))
ax[0].plot(df['New_Date'],col_mil,color='red')
ax[1].plot(df['New_Date'],col_inch,color='blue')
plt.show()
```
