```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'Temp_Dec.csv',header=[0,1])
df.columns = df.columns.map('_'.join)
''' converting Fahrenheit to Celsius (45°F − 32) × 5/9'''
convert_to_cel_max_temp = (df['Temperature (°F)_Max'] - 32) * 5/9
convert_to_cel_min_temp = (df['Temperature (°F)_Min'] - 32) * 5/9

fig,aix = plt.subplots(1,2,figsize=(10, 7))
aix[1].set_title('Temperature')
aix[1].set_xticks(df['Time_Dec']) 
aix[1].set_xlabel('Date')
aix[1].set_ylabel('Celsius')
aix[1].plot(df['Time_Dec'],convert_to_cel_max_temp,label="Max")
aix[1].plot(df['Time_Dec'],convert_to_cel_min_temp,label="Min")
aix[0].set_title('Humidity')
aix[0].set_xticks(df['Time_Dec']) 
aix[0].set_xlabel('Date')
aix[0].set_ylabel('Humidity')
aix[0].plot(df['Time_Dec'],df['Humidity (%)_Max'])
aix[0].plot(df['Time_Dec'],df['Humidity (%)_Min'])
plt.show()
print(df.head(0))
```
